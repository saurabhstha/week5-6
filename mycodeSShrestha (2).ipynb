```python
#import required libraries
import pandas as pd
import numpy as np
```

```python
# load the data
df = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')
df
```

## Chapter 7
* Filter out missing data
* Remove duplicates

```python
# Filter missing data.
df.dropna(thresh=18, inplace = True)
df
```

```python
# Remove duplicates
df.drop_duplicates()
df
```

## Chapter 8

*Create hierarchical index  
*Reshape  
*Pivot the data  

```python
# using the pandas columns attribute.
col = df.columns
print(col)
```

```python
# using the pandas set_index() function.
df_ind3 = df.set_index([ 'Your gender:','How old are you?', 'Which country do you live in?'])
  
# sort the data by using sort_index()
df_ind3.sort_index()
  
print(df_ind3.head(10))
```

```python
# reshape to rearrange data in a DataFrame.

df1 = df.stack()
df1
```

```python
# pivots from the rows into the columns

df1.unstack()
```

```python
# unstack a different level by passing a level number or name:

df1.unstack(0)
```

## Chapter 10  
* Grouping with Dicts/Series

```python

# applying groupby() function to
# group the data on team value.
gk = df.groupby('Your gender:')
  
# Let's print the first entries
# in all the groups formed.
gk.first()
```

```python
# Finding the values contained in the "Male" group
gk.get_group('Male')
```

```python
# Cross Tabs
pd.crosstab(df['Your gender:'], df['How old are you?'], margins=True)
```
